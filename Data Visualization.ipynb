```python
flight_data=pd.read_csv(r"C:\Users\HP\Documents\Python\Dataset\data-for-datavis\flight_delays.csv",index_col='Month')
```

```python
flight_data.head()
```

```python
plt.figure(figsize=(16,6))
plt.title("Average arrival delay for Inceptez Airlines WNm by month")

#definition for the bar
sns.barplot(x=flight_data.index,y=flight_data['WN'])

#Adding lablels
plt.xlabel("Arrival delay in mins")
plt.ylabel("The months as integers")
```

```python
#Try for NK
plt.figure(figsize=(16,6))
plt.title("Average arrival delay for Inceptez Airlines WNm by month")

#definition for the bar
sns.barplot(x=flight_data.index,y=flight_data['NK'])

#Adding lablels
plt.xlabel("Arrival delay in mins")
plt.ylabel("The months as integers")
```

```python
plt.figure(figsize=(8,8))
plt.title("Average arrival delay for Inceptez Airlines WNm by month")

#definition for the heatmap
sns.heatmap(data=flight_data,annot=True)

#Adding lablels
plt.xlabel("Arrival delay in mins")
plt.ylabel("The months as integers")
```

```python
plt.figure(figsize=(8,8))
plt.title("Average arrival delay for Inceptez Airlines WNm by month")

#definition for the heatmap
sns.heatmap(data=flight_data,annot=True,vmin=-10,vmax=-10,cmap='coolwarm')


#Adding lablels
plt.xlabel("Arrival delay in mins")
plt.ylabel("The months as integers")
```
